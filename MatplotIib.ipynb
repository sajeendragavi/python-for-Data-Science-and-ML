```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel(' X Label')

plt.ylabel('Y label')
plt.title('Title')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#object oriented method
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Set Title')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
axes[0].plot(x,y)
axes[0].set_title('First Plot')
```

```python
axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='X Squared',color = 'purple',linewidth = 1,linestyle = '-',marker = '.')
ax.plot(x,x**3,label='X Cubed')

ax.legend(loc=10)

```

NOTE:All the commands for plotting a figure should all go in the same cell.seperating them out into multiple cells may cause nothing to show up.
